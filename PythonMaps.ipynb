```python
url = geodatasets.get_path('geoda.nyc_neighborhoods')

df = gpd.read_file(url)

print(df.head())

df.plot()
```

```python
import pandas as pd
df['popdty'] = pd.to_numeric(df['popdty'], errors='coerce')  # Convert to float
```

```python
df['popdty'] = df['popdty'].round(0)  # Round after conversion

df.plot(column='popdty', legend=True)
```

```python
#Certain categorical columns can be visualised easily:
df.plot(column='boroname', cmap='Set3', legend=True)
```

```python
#You can also visualise statistics by borough. Try the following:


borough_avg = df.groupby('boroname')['popdty'].mean().to_dict()

df['borough_popdty'] = df['boroname'].map(borough_avg)

df.plot(column='borough_popdty', cmap='OrRd', legend=True, edgecolor=None)

#A categorical column map of NYC neighbourhoods categorised by borough.
```

```python
#an alternative visualisation like a bar chart could be an better approach
import seaborn as sns
sns.barplot(

    x=df.groupby('boroname')['popdty'].mean().sort_values().index,

    y=df.groupby('boroname')['popdty'].mean().sort_values().values

).set(title="Average Population Density by Borough")
```

```python
#Let’s say you want to pinpoint a particular location. You could do that using a red dot. If you know the longitude and latitude, you can visualise it.
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

df.plot( cmap=None, legend=True, edgecolor='black', ax=ax)

ax.scatter(-73.9442, 40.8075, color='red', s=100, label="I was here!", alpha=0.7)

ax.scatter(-74.0445, 40.6892, color='blue', s=100, label="Statue of Liberty", alpha=0.7)

ax.scatter(-73.9683, 40.7851, color='orange', s=100, label="Central Park", alpha=0.7)

ax.scatter(-73.985, 40.7488, color='white', s=100, label="Empire State building", alpha=0.7)

ax.scatter(-73.9857, 40.7580, color='pink', s=100, label="Time Square", alpha=0.7)

ax.set_title("NYC Population Density with Highlighted Point", fontsize=14)

ax.legend()

plt.show()
```
